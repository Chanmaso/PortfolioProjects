```python
# First let's import the packages that will be used in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# to read in the data
df = pd.read_csv(r'C:\Users\Chanmaso R Shimray\Desktop\Data analyst portfolios dataset\movies.csv')
```

```python
df.head()
```

```python
# checking the missing data if any using a loop

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# column`s data types

print(df.dtypes)
```

```python
# changing the data types

df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
# correcting a year column
df['yearcorrect']=df['released'].str.extract(r', (\d{4})')
df
```

```python
dfs=df.sort_values(by=['gross'], inplace = False, ascending = False)
pd.set_option('display.max_rows', None)
dfs.head()
```

```python
# drop any duplicate

dfs.drop_duplicates().head()
```

```python
# scatter plot with budget vs gross
plt.figure(figsize=(10,6))
plt.scatter(x=dfs['budget'], y=dfs['gross'])
plt.title ('Budget vs Gross Earning',fontsize=16, weight='bold')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for film')
plt.show()
```

```python
# using seaborn

sns.regplot(x='budget', y='gross', data=dfs, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# using correlation\
dfs=df.select_dtypes(include=[np.number]) # converting to numeric

dfs.corr(method='spearman') #pearson , kendall, spearman
```

```python
# high correlation betn budget and gross
plt.figure(figsize=(10,6))
correlation_matrix=dfs.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title ('Correlation matrix for numeric features',fontsize=16, weight='bold')
plt.xlabel('movie Features')
plt.ylabel('movie Features')
plt.show()
```

```python
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
        
```

```python
plt.figure(figsize=(12,8))
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title ('Correlation matrix for numeric features',fontsize=16, weight='bold')
plt.xlabel('movie Features')
plt.ylabel('movie Features')
plt.show()
```
